```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as warnings
warnings.filterwarnings("ignore")
```

```python
mba_df = pd.read_csv('mba_decision_dataset.csv')
mba_df
```

```python
gender= mba_df['Gender'].value_counts()
```

```python
plt.figure(figsize=(8,8))
colors = sns.color_palette('pastel')[0:len(gender)]
plt.pie(gender, labels=gender.index, autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Gender Distibution')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=mba_df, x='Age', bins=10, kde=True, color='skyblue')

plt.title('Distribution of Age', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

plt.show()
              
```

```python
plt.figure(figsize=(10,6))

sns.countplot(y='Undergraduate Major', data=mba_df, 
              order=mba_df['Undergraduate Major'].value_counts().index, palette='pastel', hue='Undergraduate Major', legend=False)
plt.title('Distribution of Undergraduate Major', fontsize=14)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Undergraduate Major', fontsize=12)
plt.show
```

```python
plt.figure(figsize=(10,6))

sns.countplot(y='Location Preference (Post-MBA)', data=mba_df, order=mba_df['Location Preference (Post-MBA)'].value_counts().index, palette='viridis', hue='Location Preference (Post-MBA)', legend=False)
plt.title('Distribution of Location Preference (Post-MBA)', fontsize=14)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Location Preference', fontsize=12)
plt.show()
```
